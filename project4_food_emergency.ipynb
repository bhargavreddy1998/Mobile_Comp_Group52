```python
import requests
import re
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/List_of_emergency_telephone_numbers"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

tables = soup.find_all('table')
ambulance_codes = {}

for table in tables:
  rows = table.find_all('tr')
  for row in rows:
      cells = row.find_all('td')
      if len(cells) > 1:
          html_snippet = cells[0]
          country_name = html_snippet.find('a').get_text(strip=True)
          ambulance_number = cells[1].get_text(strip=True)
          ambulance_number = re.sub(r'\[\d+\]', '', ambulance_number)
          ambulance_codes[country_name] = ambulance_number
print(ambulance_codes)
```
